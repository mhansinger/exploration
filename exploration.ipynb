### Berechnung mit Methoden 

```python
def get_interval_indices(
    signals: pd.DataFrame, start_time_stamp: pd.Timestamp
) -> tuple[pd.Timestamp, pd.Timestamp, pd.Timestamp]:
    """extracts the indices (timestamps) 10000 sec. before and 10000 sec. after the start of Speed_B signal."""

    # idx where Speed B strats
    signal_B_start_idx = signals.speed_B.loc[start_time_stamp].first_valid_index()

    idx_min = signal_B_start_idx - pd.Timedelta(10000, "s")
    idx_max = signal_B_start_idx + pd.Timedelta(10000, "s")

    return idx_min, idx_max, signal_B_start_idx
```

```python
def get_lag(signals_subset: pd.DataFrame) -> int:
    """cross correlate Speed A and Speed B
    Args:
        signals_subset (pd.DataFrame): subset of resampled Speed A and Speed B.
    Returns:
        int: index where in the intervall [-10000:10000] maximum correlation appears.
    """

    corr = correlate(
        signals_subset.speed_A,
        signals_subset.speed_B.dropna(),
        mode="valid",
        method="direct",
    )

    max_corr_index = np.argmax(corr)

    return max_corr_index
```

```python
# alles als Methode zusammenpacken


def get_time_shift_correlate(signals: pd.DataFrame, start_search_date: str):
    """Berechnugn des Lags / shift

    Args:
        signals (pd.DataFrame): _description_
        start_search_date (str): _description_

    Raises:
        ValueError: _description_
    """

    allowed_start_dates = {"2018-12-02", "2018-12-03", "2018-12-04", "2018-12-05"}
    if start_search_date not in allowed_start_dates:
        raise ValueError(
            f"Invalid: {start_search_date}. Allowed dates: {allowed_start_dates}"
        )

    # Start and end index of the interval around Speed B signal
    idx_min, idx_max, signal_B_start_idx = get_interval_indices(
        signals=signals, start_time_stamp=start_search_date
    )

    # Signale A und B "freischneiden"
    signals_subset = signals.loc[idx_min:idx_max]

    lag = get_lag(signals_subset=signals_subset)

    print(
        f"Ausgehend vom Start des Intervalls liegt der optimale Versatz bei {lag} Sekunden."
    )

    # Index, bei dem Speed_B am besten mit Speed_A übereinstimmt
    idx_ideal_overlap = idx_min + pd.Timedelta(lag, "s")
    time_shift = idx_ideal_overlap - signal_B_start_idx

    print(
        f"Der Zeitversatz beträgt {time_shift.total_seconds()} Sekunden, bezogen auf die Uhrzeit von Signal B."
    )
    print(f"Signalstart B: {signal_B_start_idx}")
    print(f"optimaler Start: {idx_ideal_overlap}\n")

    # Plotting
    plt.figure(figsize=(15, 1))
    plt.title(f"Vergleich der Signale am {start_search_date}")
    signals.speed_A.loc[idx_min:idx_max].plot()
    signals.speed_B.loc[idx_min:idx_max].shift(int(time_shift.total_seconds())).plot()
    signals.speed_B.loc[idx_min:idx_max].plot()
    plt.legend(["Signal A", "Signal B shifted", "Signal B (eigene Zeit)"])
```

### 1.1 Verwendung von normalisierten Werten


```python
def get_lag_norm(signals_subset: pd.DataFrame) -> int:
    """cross correlate Speed A and Speed B with signals normalized on the respective interval.
    Args:
        signals_subset (pd.DataFrame): subset of resampled Speed A and Speed B.
    Returns:
        int: index where in the intervall [-10000:10000] maximum correlation appears.
    """

    print("Mit Normalisierung der Signale")

    speed_A_norm = (
        signals_subset.speed_A - signals_subset.speed_A.mean()
    ) / signals_subset.speed_A.std()
    speed_B_norm = (
        signals_subset.speed_B.dropna() - signals_subset.speed_B.dropna().mean()
    ) / signals_subset.speed_B.dropna().std()

    corr = correlate(
        speed_A_norm,
        speed_B_norm,
        mode="valid",
        method="direct",
    )

    max_corr_index = np.argmax(corr)

    return max_corr_index
```

```python
def get_time_shift_correlate_norm(signals: pd.DataFrame, start_search_date: str):
    """Berechnugn des Lags / shift

    Args:
        signals (pd.DataFrame): _description_
        start_search_date (str): _description_

    Raises:
        ValueError: _description_
    """

    allowed_start_dates = {"2018-12-02", "2018-12-03", "2018-12-04", "2018-12-05"}
    if start_search_date not in allowed_start_dates:
        raise ValueError(
            f"Invalid: {start_search_date}. Allowed dates: {allowed_start_dates}"
        )

    # Start and end index of the interval around Speed B signal
    idx_min, idx_max, signal_B_start_idx = get_interval_indices(
        signals=signals, start_time_stamp=start_search_date
    )

    # Signale A und B "freischneiden"
    signals_subset = signals.loc[idx_min:idx_max]

    lag = get_lag_norm(signals_subset=signals_subset)

    print(
        f"Ausgehend vom Start des Intervalls liegt der optimale Versatz bei {lag} Sekunden."
    )

    # Index, bei dem Speed_B am besten mit Speed_A übereinstimmt
    idx_ideal_overlap = idx_min + pd.Timedelta(lag, "s")
    time_shift = idx_ideal_overlap - signal_B_start_idx

    print(
        f"Der Zeitversatz beträgt {time_shift.total_seconds()} Sekunden, bezogen auf die Uhrzeit von Signal B."
    )
    print(f"Signalstart B: {signal_B_start_idx}")
    print(f"optimaler Start: {idx_ideal_overlap}\n")

    # Plotting
    plt.figure(figsize=(15, 1))
    plt.title(f"Vergleich der Signale am {start_search_date}")
    signals.speed_A.loc[idx_min:idx_max].plot()
    signals.speed_B.loc[idx_min:idx_max].shift(int(time_shift.total_seconds())).plot()
    signals.speed_B.loc[idx_min:idx_max].plot()
    plt.legend(["Signal A", "Signal B shifted", "Signal B (eigene Zeit)"])
```

# 2. Versatz 'händisch' berechen
Schiebe Signal von Speed B im Intervall von -10000 bis 10000 sec. über Signal von Speed A. <br>
Berechne die Abweichung/Differenz zwischen Speed B und Speed A (auf der Länge von Speed B). <br>
Bei der ID mit der kleinsten Abweichung sollten die Signale identisch sein.

```python
def shift_compare_signals(
    speed_A: np.ndarray, speed_B: np.ndarray
) -> tuple[int, float]:
    """shifts gradually the signal of Speed B over Speed A and computes the residual
    between the two signals over the length of signal B.

    Args:
        speed_A (np.ndarray): _description_
        speed_B (np.ndarray): _description_

    Returns:
        tuple[int, float]: id of best location, minimum average residual
    """

    len_A = len(speed_A)
    len_B = len(speed_B)

    best_shift_id = 0
    min_residual = 1e6

    for i in range(len_A - len_B):
        residual = abs(speed_A[i : i + len_B] - speed_B).sum()

        if residual < min_residual:
            min_residual = residual
            best_shift_id = i

    return best_shift_id, (min_residual / len_B)
```

```python
# alles als Methode zusammenpacken


def get_time_shift_iterate(signals: pd.DataFrame, start_search_date: str):

    allowed_start_dates = {"2018-12-02", "2018-12-03", "2018-12-04", "2018-12-05"}
    if start_search_date not in allowed_start_dates:
        raise ValueError(
            f"Ungültiger Wert: {start_search_date}. Erlaubte Werte sind: {allowed_start_dates}"
        )

    # Indizes des zu betrachtenden Intervalls
    idx_min, idx_max, signal_B_start_idx = get_interval_indices(
        signals=signals, start_time_stamp=start_search_date
    )

    # Signale A und B "freischneiden"
    signals_subset = signals.loc[idx_min:idx_max]

    speed_A_np = signals_subset.speed_A.to_numpy()
    speed_B_np = signals_subset.speed_B.dropna().to_numpy()  # drop NaNs

    best_shift_id, avg_residual = shift_compare_signals(
        speed_A=speed_A_np, speed_B=speed_B_np
    )
    print(
        f"Ausgehend vom Start des Intervalls liegt der optimale Versatz bei {best_shift_id} Sekunden."
    )

    # Index, bei dem Speed_B am besten mit Speed_A übereinstimmt
    idx_ideal_overlap = idx_min + pd.Timedelta(best_shift_id, "s")
    time_shift = idx_ideal_overlap - signal_B_start_idx

    print(
        f"Der Zeitversatz beträgt {time_shift.total_seconds()} Sekunden, bezogen auf die Uhrzeit von Signal B."
    )
    print(f"Durchschnittlicher Fehler {avg_residual:.3f} m/s.")
    print(f"Signalstart B: {signal_B_start_idx}")
    print(f"optimaler Start: {idx_ideal_overlap}\n")

    # Plotting
    plt.figure(figsize=(15, 1))
    plt.title(f"Vergleich der Signale am {start_search_date}")
    signals.speed_A.loc[idx_min:idx_max].plot()
    signals.speed_B.loc[idx_min:idx_max].shift(int(time_shift.total_seconds())).plot()
    signals.speed_B.loc[idx_min:idx_max].plot()
    plt.legend(["Signal A", "Signal B shifted", "Signal B (eigene Zeit)"])
```
